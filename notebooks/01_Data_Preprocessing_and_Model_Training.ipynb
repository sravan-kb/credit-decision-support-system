```python
replace_zero_cols = [
    "CC_utilization",
    "PL_utilization",
    "time_since_first_deliquency",
    "time_since_recent_deliquency",
    "max_delinquency_level",
    "max_deliq_6mts",
    "max_deliq_12mts",
    "tot_enq",
    "CC_enq",
    "CC_enq_L6m",
    "CC_enq_L12m",
    "PL_enq",
    "PL_enq_L6m",
    "PL_enq_L12m",
    "enq_L12m",
    "enq_L6m",
    "enq_L3m"
]

df_clean[replace_zero_cols] = df_clean[replace_zero_cols].replace(-99999, 0)
```

```python
import numpy as np

replace_nan_cols = [
    "max_unsec_exposure_inPct",
    "time_since_recent_enq",
    "time_since_recent_payment",
    "pct_currentBal_all_TL"
]

df_clean[replace_nan_cols] = df_clean[replace_nan_cols].replace(-99999, np.nan)
```

```python
numeric_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
import matplotlib.pyplot as plt

approval_counts = df_clean["Approved_Flag"].value_counts()

plt.figure(figsize=(7,5))

bars= plt.bar(
    approval_counts.index,
    approval_counts.values,
    edgecolor="black"
)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 300,
        f"{int(bar.get_height())}",
        ha="center"
    )

plt.title("Distribution of Approved Flag")
plt.xlabel("Approval Category")
plt.ylabel("Number of Customers")
plt.show()
```
